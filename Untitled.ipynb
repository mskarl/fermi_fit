```python
factor = 3
MeV_to_erg = 1.60218e-6
ul_ts_threshold = 4

def plot_sed(sed):
    m = sed['ts'] < ul_ts_threshold
    x = sed['e_ctr']#/1e3
    y = sed['e2dnde']#*MeV_to_erg
    yerr = sed['e2dnde_err']#*MeV_to_erg
    yerr_lo = sed['e2dnde_err_lo']#*MeV_to_erg
    yerr_hi = sed['e2dnde_err_hi']#*MeV_to_erg
    yul = sed['e2dnde_ul95']#*MeV_to_erg
    delo = (sed['e_ctr'] - sed['e_min'])/1e3
    dehi = (sed['e_max'] - sed['e_ctr'])/1e3
    xerr0 = np.vstack((delo[m], dehi[m]))
    xerr1 = np.vstack((delo[~m], dehi[~m]))
    
    plt.errorbar(x[~m], y[~m], xerr=xerr1,
                yerr=(yerr_lo[~m], yerr_hi[~m]),
                linestyle='', fmt='o')
    
    serr = 10**np.log10(yul[m]) - 10**(np.log10(yul[m])-0.1/8.*factor)
    plt.errorbar(x[m], yul[m], xerr=xerr0,
                yerr=serr,  uplims=True, fmt='o')
    plt.loglog()

    
def plot_sed_theo(sed):
    m = sed[b'ts'] < ul_ts_threshold
    x = sed[b'e_ctr']#/1e3
    y = sed[b'e2dnde']#*MeV_to_erg
    yerr = sed[b'e2dnde_err']#*MeV_to_erg
    yerr_lo = sed[b'e2dnde_err_lo']#*MeV_to_erg
    yerr_hi = sed[b'e2dnde_err_hi']#*MeV_to_erg
    yul = sed[b'e2dnde_ul95']#*MeV_to_erg
    delo = (sed[b'e_ctr'] - sed[b'e_min'])/1e3
    dehi = (sed[b'e_max'] - sed[b'e_ctr'])/1e3
    xerr0 = np.vstack((delo[m], dehi[m]))
    xerr1 = np.vstack((delo[~m], dehi[~m]))
    
    plt.errorbar(x[~m], y[~m], xerr=xerr1,
                yerr=(yerr_lo[~m], yerr_hi[~m]),
                linestyle='', fmt='o')
    
    serr = 10**np.log10(yul[m]) - 10**(np.log10(yul[m])-0.1/8.*factor)
    plt.errorbar(x[m], yul[m], xerr=xerr0,
                yerr=serr,  uplims=True, fmt='o')
    plt.loglog()
    
    
def plot_llh_sed(c, source_name):
    E = np.array(c['sources'][source_name]['model_flux']['energies'])#/1e3
    dnde = np.array(c['sources'][source_name]['model_flux']['dnde'])#*MeV_to_erg
    dnde_hi = np.array(c['sources'][source_name]['model_flux']['dnde_hi'])#*MeV_to_erg
    dnde_lo = np.array(c['sources'][source_name]['model_flux']['dnde_lo'])#*MeV_to_erg
    
    plt.plot(E, (E**2)*dnde, 'k--')
    plt.plot(E, (E**2)*dnde_hi, 'k')
    plt.plot(E, (E**2)*dnde_lo, 'k')
    
    

    
def plot_llh_sed_theo(c, source_name):
    E = np.array(c[b'sources'][source_name][b'model_flux'][b'energies'])#/1e3
    dnde = np.array(c[b'sources'][source_name][b'model_flux'][b'dnde'])#*MeV_to_erg
    dnde_hi = np.array(c[b'sources'][source_name][b'model_flux'][b'dnde_hi'])#*MeV_to_erg
    dnde_lo = np.array(c[b'sources'][source_name][b'model_flux'][b'dnde_lo'])#*MeV_to_erg
    
    plt.plot(E, (E**2)*dnde, 'k--')
    plt.plot(E, (E**2)*dnde_hi, 'k')
    plt.plot(E, (E**2)*dnde_lo, 'k')


def plot_bowtie(bowtie, bowtie_fill=False):
    energies = np.array(10 ** bowtie['log_energies'])#/1e3
    e2 = 10 ** (2 * bowtie['log_energies'])
    factor = e2 #* MeV_to_erg
    plt.plot(energies,
            bowtie['dnde'] * factor, color="grey", linestyle='--')
    plt.plot(energies,
            bowtie['dnde_lo'] * factor, color="grey")
    plt.plot(energies,
            bowtie['dnde_hi'] * factor, color="grey")
    if bowtie_fill:
        plt.fill_between(energies,
                        bowtie['dnde_lo'] * factor,
                        bowtie['dnde_hi'] * factor,
                        alpha=0.5)
    
    
    
def plot_bowtie_theo(bowtie, bowtie_fill=False):
    energies = np.array(10 ** bowtie[b'log_energies'])#/1e3
    e2 = 10 ** (2 * bowtie[b'log_energies'])
    factor = e2 #* MeV_to_erg
    plt.plot(energies,
            bowtie[b'dnde'] * factor, color="black", linestyle='--')
    plt.plot(energies,
            bowtie[b'dnde_lo'] * factor, color="black")
    plt.plot(energies,
            bowtie[b'dnde_hi'] * factor, color="black")
    if bowtie_fill:
        plt.fill_between(energies,
                        bowtie[b'dnde_lo'] * factor,
                        bowtie[b'dnde_hi'] * factor,
                        alpha=0.5)
    
```
